# Librerías

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

# Def. Variables

```python
h=0.01
x=np.linspace(0.1,1.1,100)
def f(x):
    return np.sqrt(np.tan(x))
```

# Derivadas

```python
class Derivate():
    def __init__(self,x,f):
        self.x=x
        self.f=f
    
    def progre(self):
        return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))
    def central(self):
        return (f(x+h)-f(x-h))/(2*h)
    def real_value(self):
        return (1/(np.cos(x))**2) / (2*np.sqrt(np.tan(x)))

d=Derivate(x,f)
```

```python
central_d = d.central()
progre_d = d.progre()
real_value_d = d.real_value()
```

### Gráficas

```python
plt.plot(x, central_d, color='g')
plt.title("Central")
plt.xlabel("$\it{x}$")
plt.ylabel("$\it{y}$")
plt.show()

plt.plot(x, progre_d, color = 'b')
plt.title("Progre.")
plt.xlabel("$\it{x}$")
plt.ylabel("$\it{y}$")
plt.show()

plt.plot(x, real_value_d, color = 'r')
plt.title("Valor real")
plt.xlabel("$\it{x}$")
plt.ylabel("$\it{y}$")
plt.show()
```

# Errores

```python
error_central = np.abs(central_d-real_value_d)
error_progre = np.abs(progre_d-real_value_d)
```

### Gráficas

```python
fig = plt.figure(figsize=(15, 4))
ax = fig.add_subplot(1, 2, 1)
ax.plot(x, error_central, color='g')
ax.set_title("Central")
ax.set_xlabel("$\it{x}$")
ax.set_ylabel("Error")
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x, error_progre, color = 'b')
ax2.set_title("Progre.")
ax2.set_xlabel("$\it{x}$")
ax2.set_ylabel("Error")

plt.suptitle("Errores locales",fontsize=20)
plt.show()
```
