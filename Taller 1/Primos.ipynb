```python
import numpy as np
import matplotlib.pyplot as plt
```

# Buscador de primos

```python
def primo(n):
  for i in range(2,int(((n**(1/2)))+1)):
    if (n%i) == 0:
      return False
  return True
```

```python
def mil_primos():
    list=[]
    n=2

    while len(list)<1000:
        if primo(n):
            list.append(n)
        n+=1
    return list
```

# Parte visual

```python
# ==============================
# Gr√°fica
# ==============================
coor_x = np.arange(1,1001)
coor_y = mil_primos()
plt.plot(coor_x,coor_y, "-r")
plt.show()

# ==============================
# Primeros 10 primos
# ==============================
print("Los primeros 10 n√∫meros primos son:")
for i in range(10):
    print("\u2022", coor_y[i])
```
