```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_sides=5
A = 1
t = np.linspace(0,(2*np.pi),200)
n_x = [1,2,3,4,5]
delta = [0,np.pi/4,np.pi/2]
fig = plt.figure(figsize=(5,6))
```

```python
def lissajous(delta):
    k=1
    for w_x in range(1,len(n_x)):
        x = A*np.sin(w_x*t)
        for w_y in range(w_x,len(n_x)):
            y = A*np.sin(w_y*t + delta)
            plt.subplot(5,5,k)
            k += 1
            plt.plot(x,y)
            plt.axis("off")
    plt.show()

for i in range(3):
    lissajous(delta[i])
```
