# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
```

# Lectura del archivo

```python
datos = open("EstrellaEspectro.txt")
```

# Def. variables

```python
maximos_x = []
maximos_y = []
x=[]
y=[]
valor_anterior = [0,0]
valor = [0,0]
```

# Asignación de coordenadas

```python
for linea in datos:
    valor_siguiente = linea.split("  ")
    x.append(float(valor[0]))
    y.append(float(valor[1]))
    if float(valor_siguiente[1]) < float(valor[1]) and float(valor[1])>float(valor_anterior[1]):
        maximos_x.append(float(valor[0]))
        maximos_y.append(float(valor[1]))
    valor_anterior = valor
    valor = valor_siguiente
```

# Gráficas

```python
plt.plot(x[1:],y[1:])
plt.scatter(maximos_x,maximos_y,c='r', s=10)
plt.xlabel("$\it{x}$")
plt.ylabel("$\it{y}$")
plt.title("Grafico de el espectro de estrellas")
plt.show()
```
