# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Discretización

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
X,Y = np.meshgrid(x,y)
h = 0.001
```

 # Función potencial

```python
def Potencial(x,y):
    return (2*x*(1-(4/(x**2+y**2))))
```

# Derivadas

```python
PartialOperator = (lambda f,x,y : (f(x+h,y)-f(x-h,y))/(2*h),\
           lambda f,x,y : -(f(x,y+h)-f(x,y-h))/(2*h))
```

```python
Ex = PartialOperator[0](Potencial,X,Y)
Ey = PartialOperator[1](Potencial,X,Y)
Ex = Ex.T
Ey = Ey.T
```

# Gráfica

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)

for i in range(1,len(x)):
    for j in range(1,len(y)):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j],color='b')
```
