# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
```

 # Función potencial

```python
def Potencial(x,y):
    return (2*x*(1-(4/(x**2+y**2))))
```

# Derivadas

```python
PartialOperator = (lambda f,x,y,h : (f(x+h,y)-f(x-h,y))/(2*h),\
           lambda f,x,y,h : -(f(x,y+h)-f(x,y-h))/(2*h))
```

# Discretización

```python
n=25
h = 0.001
x = np.linspace(-4,4,n)
y = np.linspace(-4,4,n)

Ex = np.empty((n,n))
Ey = np.empty((n,n))

for i in range(1,len(x)):
    for j in range(1,len(x)):
        if (x[i]**2 + y[j]**2) > 4:
            Ex[i,j] = PartialOperator[0](Potencial,x[i], y[j],h)
            Ey[i,j] = PartialOperator[1](Potencial,x[i], y[j],h)

        else:
            Ex[i,j] = 0
            Ey[i,j] = 0
```

# Gráfica

```python
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel("$\it{x}$ (cm)")
ax.set_ylabel("$\it{y}$ (cm)")

ax.plot(2*np.cos(np.linspace(0, 2*np.pi)), 2*np.sin(np.linspace(0, 2*np.pi)),color='r')

for i in range(1,len(x)):
    for j in range(1,len(x)):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j],color='b')

plt.show()
```
