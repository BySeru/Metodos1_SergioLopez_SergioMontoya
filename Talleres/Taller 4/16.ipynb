# Librerías

```python
import numpy as np
import sympy as sp
import legendre as lg
import matplotlib.pyplot as plt
```

# Def. Variables

```python
xi = np.linspace(-1,1,100)
def f(x):
    return np.exp(x)*(x**3/(np.exp(x)-1))

a = 0
b = 1
n = 3
```

# Legendre

```python
Roots,weights=np.polynomial.laguerre.laggauss(n)
```

# Valor de la integral

```python
int2 = np.sum(weights*f(Roots))
print("Siendo n=3: ", round(int2,6))
```

# Gráfica y error

```python
error = lambda I: (I)/((np.pi**4)/15)
x= np.arange(2,11)
y=[]
for i in x:
    roots,weights=np.polynomial.laguerre.laggauss(i)
    int = np.sum(weights*f(roots))
    y.append(error(int))

plt.axhline(1,color="k",linestyle="dashed")
plt.scatter(x,y, label="Laguerre quadrature acurracy")
plt.legend()
plt.grid()
plt.show()
```
