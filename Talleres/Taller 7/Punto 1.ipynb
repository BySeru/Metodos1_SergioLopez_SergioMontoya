# Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

# Definición de matrices

```python
A = np.array([[2, -1],[1, 2], [1, 1]])
b = np.array([[2], [1], [4]])
```

# Solución Minimos cuadrados

```python
def GetFit(A, b):
    AT = np.dot(A.T,A)
    bT = A.T @ b

    xsol = np.linalg.solve(AT,bT)
    
    return xsol
sol = GetFit(A,b)
sol
```

# Grafica

Para graficar se despejo y de las ecuaciones originales y se grafico junto a x. Esto nos queda
\begin{align}
&2x-2 = y\\
&\frac{x-1}{-2}=y \rightarrow \frac{1-x}{2} = y\\
&4-x = y
\end{align}

```python
x = np.arange(-5,5,0.01)
plt.plot(x, 2*x-2)
plt.plot(x, (1-x)/2)
plt.plot(x, 4-x)
plt.plot(sol[0,0], sol[1,0], marker='o', color='b')
```

```python
x = np.arange(-5,5,0.01)
y = np.arange(-5,5,0.01)
X,Y = np.meshgrid(x, y)
dist = np.zeros((np.shape(x)[0], np.shape(y)[0]))
for i in range(x.size):
    for j in range(y.size):
        dist[i,j] = np.linalg.norm(np.dot(A,[[x[i]],[y[j]]])-b)
dist
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))
surf = ax.plot_surface(X, Y, dist, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$d(x^*)$')
ax.view_init(30, 50)
fig.colorbar(surf, orientation='vertical', label = 'distance')
```

```python
xm, ym = np.where(dist == np.min(dist))
print('El punto minimo es ', '(',x[xm],y[ym],')')
```
