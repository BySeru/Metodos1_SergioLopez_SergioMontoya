# Librerías

```python
import yaml
import numpy as np
import sympy as sp
import Newt_Raph as nr
```

# Def. Polinomio de Laguerre

```python
def Laguerre(n):
    x = sp.Symbol("x",Real=True)
    y = sp.Symbol("y",Real=True)
    
    poly = ((np.e**x)/np.math.factorial(n))*sp.diff(((np.e**(-x))*(x**n)),x,n)
    poly=sp.lambdify([x],poly,'numpy')
    return poly
```

# Raíces hasta el vigésimo polinomio

```python
def roots_20p():
    roots={}
    
    for i in range(1,21):
        x = np.linspace(0,200,100)
        r = nr.GetAllRoots(x,Laguerre(i))
            
        roots["Polinomio "+str(i)] = r.tolist()
        
    return roots
```

# Presentación de las raíces

```python
roots=roots_20p()
print(yaml.dump(roots, sort_keys=False, default_flow_style=False))
```
