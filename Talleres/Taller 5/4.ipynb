# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Def. Variables

```python
N = int(10**4)
x = np.random.uniform(0,100,N)
corre = {}
```

# Coeficientes

```python
def Correlations(k):
    Sum = 0
    for i in range(N):
        Sum += (x[i]*x[i+k])
        return Sum/N

k = np.array(range(0, 30))
for i in k:
    corre[i] = Correlations(i)
```

# Gráfica

```python
coeff = list(corre.values())

plt.plot(k,coeff)
plt.grid()
plt.show()
```
