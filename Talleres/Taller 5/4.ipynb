# Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Def. Variables

```python
N = int(10**4)
x = np.random.rand(N+30)
k = np.linspace(1,30,30,dtype=int)
coeff = {}
```

# Coeficientes

```python
def c(k,x):
  x_i = (x)[0:N]
  x_ik = (x)[k:(N+k)]
  return (1/N)*np.sum(x_i*x_ik)

for i in k:
  coeff[i] = (c(i,x))
```

# Gráfica

```python
p = coeff.values()

plt.style.use("seaborn-whitegrid")
plt.ylabel('c(K)')
plt.xlabel('Puntos K')
plt.ylim((0.2,0.3))
plt.plot(k,p)
plt.show()
```
